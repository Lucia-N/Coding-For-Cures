
### The Python Interpreter as a Calculator: Basic Arithmetic Operations

One of the simplest ways to start using Python is as a calculator. You can perform basic arithmetic operations directly in the interpreter or within a Python script.

Python supports the following standard arithmetic operators:
*   `+` (Addition)
*   `-` (Subtraction)
*   `*` (Multiplication)
*   `/` (Division) - results in a float
*   `//` (Integer Division) - results in an integer, discarding the fractional part
*   `%` (Modulo) - returns the remainder of the division
*   `**` (Exponentiation) - raises the first number to the power of the second

**Order of Operations (Precedence):**
Python follows the standard mathematical order of operations (PEMDAS/BODMAS): Parentheses (or Brackets), Exponents, Multiplication and Division (from left to right), Addition and Subtraction (from left to right).

**Examples:**


```python
# Addition
print("5 + 3 =", 5 + 3) # Expected: 8

# Subtraction
print("10 - 4 =", 10 - 4) # Expected: 6

# Multiplication
print("7 * 6 =", 7 * 6) # Expected: 42

# Division (Floating-point result)
print("15 / 3 =", 15 / 3) # Expected: 5.0
print("10 / 3 =", 10 / 3) # Expected: 3.3333333333333335

# Integer Division
print("10 // 3 =", 10 // 3) # Expected: 3

# Modulo (Remainder)
print("10 % 3 =", 10 % 3) # Expected: 1

# Exponentiation
print("2 ** 3 =", 2 ** 3) # Expected: 8 (2 to the power of 3)

# Combining operations with precedence
print("5 + 2 * 3 =", 5 + 2 * 3) # Expected: 11 (Multiplication before Addition)
print("(5 + 2) * 3 =", (5 + 2) * 3) # Expected: 21 (Parentheses override precedence)

# Practical use case in biomedical informatics:
# Calculating the concentration of a solution (e.g., in moles per liter)
# Molecular weight of glucose (C6H12O6) is approximately 180.16 g/mol
mass_glucose_grams = 50 # grams
volume_solution_liters = 0.5 # liters
molecular_weight_glucose = 180.16 # g/mol

# Calculate moles of glucose
moles_glucose = mass_glucose_grams / molecular_weight_glucose
print(f"Moles of glucose: {moles_glucose:.4f} mol")

# Calculate concentration (Molarity)
concentration_molar = moles_glucose / volume_solution_liters
print(f"Concentration of glucose solution: {concentration_molar:.4f} M")

# Calculate BMI (Body Mass Index) - a simple health metric
weight_kg = 70
height_m = 1.75
bmi = weight_kg / (height_m ** 2)
print(f"Body Mass Index (BMI): {bmi:.2f}")
```

### 2. Variables and Basic Data Types

In programming, a **variable** is a named storage location that holds a value. Think of it as a labeled box where you can put different types of data. Variables allow us to store and manipulate data dynamically in our programs. When you create a variable, you give it a name, and Python allocates memory to store the value associated with that name.

#### Understanding Variables: Naming Conventions and Assignment

**Naming Conventions:**
*   Variable names must start with a letter (a-z, A-Z) or an underscore (`_`). They cannot start with a number.
*   They can contain letters, numbers, and underscores.
*   Variable names are case-sensitive (`age` is different from `Age`).
*   Avoid using Python's reserved keywords (e.g., `if`, `for`, `while`, `print`, `True`, `False`, `None`).
*   It's good practice to use descriptive names (e.g., `patient_id` instead of `x`).
*   **Snake case** (`patient_id`, `gene_sequence`) is the convention for variable names in Python.

**Assignment:**
Values are assigned to variables using the assignment operator `=`. The value on the right side of `=` is stored in the variable named on the left.

**Example:**

```python
# Variable assignment
patient_name = "Alice Smith"
patient_age = 34
is_diabetic = False
genome_size_bp = 3.2 * (10**9) # Human genome size in base pairs

print(f"Patient Name: {patient_name}")
print(f"Patient Age: {patient_age}")
print(f"Is Diabetic: {is_diabetic}")
print(f"Estimated Genome Size: {genome_size_bp} bp")
```

#### Numeric Types: Integers (`int`) and Floating-Point Numbers (`float`)

Python handles numbers using two primary types:

*   **Integers (`int`):** Whole numbers (positive, negative, or zero) without a decimal point. Examples: `10`, `-5`, `0`, `1000000`.
*   **Floating-Point Numbers (`float`):** Numbers with a decimal point or numbers expressed in exponential form. Examples: `3.14`, `-0.5`, `2.718e-3` (which is 0.002718).

Python automatically infers the type of a number when you assign it to a variable.

**Example:**



```python
# Integers
num_cells = 150000 # Number of cells in a sample
chromosome_count = 23 # Number of chromosome pairs

# Floating-point numbers
ph_level = 7.4 # pH level of blood
mutation_rate = 1.2e-5 # A scientific notation for a small mutation rate
dna_concentration = 25.75 # ng/uL

print(f"Number of cells: {num_cells}, Type: {type(num_cells)}")
print(f"Chromosome count: {chromosome_count}, Type: {type(chromosome_count)}")
print(f"Blood pH: {ph_level}, Type: {type(ph_level)}")
print(f"Mutation rate: {mutation_rate}, Type: {type(mutation_rate)}")
print(f"DNA Concentration: {dna_concentration}, Type: {type(dna_concentration)}")
```

#### Text Data: Strings (`str`)

**Strings (`str`)** are sequences of characters used to represent text. They are enclosed in single quotes (`'...'`), double quotes (`"..."`), or triple quotes (`"""..."""` for multi-line strings).

**Creation:**



```python
# Single, double, and triple quoted strings
drug_name = 'Aspirin'
gene_symbol = "BRCA1"
multiline_description = """This is a multi-line string
used for longer texts, like research paper abstracts.
It can span several lines.
"""

print(drug_name)
print(gene_symbol)
print(multiline_description)
```

**Concatenation (Joining Strings):**
You can join strings using the `+` operator.

```python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(f"Full Name: {full_name}")

protein_part1 = "ATGC"
protein_part2 = "GCTA"
full_protein_sequence = protein_part1 + protein_part2
print(f"Full Protein Sequence: {full_protein_sequence}")
```

**Basic String Methods:**
Strings come with many built-in methods that allow you to perform various operations.

*   `.upper()`: Converts all characters in the string to uppercase.
*   `.lower()`: Converts all characters in the string to lowercase.
*   `.replace(old, new)`: Replaces all occurrences of a specified substring `old` with `new`.

**Example:**

```python
dna_sequence = "atggcTacgTAG"

# Convert to uppercase
upper_sequence = dna_sequence.upper()
print(f"Uppercase DNA: {upper_sequence}")

# Convert to lowercase
lower_sequence = dna_sequence.lower()
print(f"Lowercase DNA: {lower_sequence}")

# Replace a specific base (e.g., 'T' with 'U' for RNA transcription concept)
rna_like_sequence = dna_sequence.replace('T', 'U').replace('t', 'u') # Handles both cases
print(f"RNA-like sequence: {rna_like_sequence.upper()}")

# Another replace example: correcting a typo in a sample ID
sample_id_wrong = "BIO_SAMPLE-001_v1"
sample_id_correct = sample_id_wrong.replace('_v1', '_final')
print(f"Corrected Sample ID: {sample_id_correct}")
```

#### Boolean Type (`bool`): True/False Values

**Booleans (`bool`)** represent logical values: `True` or `False`. They are fundamental for conditional logic and decision-making in programs. In Python, `True` and `False` are keywords and must be capitalized.

**Example:**

```python
is_patient_positive = True # Is patient positive for a disease?
has_genetic_marker = False # Does the patient have a specific genetic marker?
result_available = True

print(f"Patient positive: {is_patient_positive}, Type: {type(is_patient_positive)}")
print(f"Has genetic marker: {has_genetic_marker}, Type: {type(has_genetic_marker)}")

# Booleans in conditional statements (covered in Control Flow)
if is_patient_positive:
    print("Initiate treatment protocol.")
else:
    print("Monitor patient condition.")

# Comparison operators often result in booleans
age = 25
is_adult = age >= 18
print(f"Is adult: {is_adult}")
```

#### Type Conversion: `int()`, `float()`, `str()`

Sometimes you need to convert data from one type to another. Python provides built-in functions for this:

*   `int()`: Converts to an integer (truncates decimal part if converting from float).
*   `float()`: Converts to a floating-point number.
*   `str()`: Converts to a string.

**Example:**

```python
# Converting between numeric types
patient_weight_str = "75.5" # Weight obtained as a string from a form
patient_weight_float = float(patient_weight_str)
patient_weight_int = int(patient_weight_float) # This will truncate the decimal part

print(f"String weight: {patient_weight_str}, Type: {type(patient_weight_str)}")
print(f"Float weight: {patient_weight_float}, Type: {type(patient_weight_float)}")
print(f"Integer weight (truncated): {patient_weight_int}, Type: {type(patient_weight_int)}")

# Converting numbers to strings (useful for displaying mixed data)
data_point = 123.45
label = "Value: " + str(data_point) # Cannot concatenate string and float directly
print(f"Combined label: {label}")

# Example: Converting a patient ID number to a string for concatenation
patient_id_num = 101
hospital_prefix = "HOS_"
full_patient_id = hospital_prefix + str(patient_id_num)
print(f"Full Patient ID: {full_patient_id}")

# Converting a boolean to an integer (True becomes 1, False becomes 0)
active_drug_status = True
status_code = int(active_drug_status)
print(f"Active drug status as integer: {status_code}")

# Be careful with conversions that might lose information or cause errors
# invalid_num_str = "abc"
# try:
#     int(invalid_num_str)
# except ValueError as e:
#     print(f"Error converting 'abc' to int: {e}")
```

Python offers several built-in data structures that are essential for organizing and managing data. #### Lists: Ordered, Mutable Sequences

A **list** is an ordered collection of items. It is one of the most versatile data structures in Python, capable of holding items of different data types (though often used for homogeneous types in scientific contexts). Lists are *mutable*, meaning their elements can be changed after creation.

**Creation:**
Lists are created by placing all the items (elements) inside square brackets `[]`, separated by commas.

**Example:**

```python
# A list of patient IDs
patient_ids = ["P001", "P002", "P003", "P004"]

# A list of gene expression values
gene_expression_values = [10.5, 12.1, 8.9, 15.3, 7.2]

# A list mixing data types (less common in scientific data, but possible)
mixed_data = ["Sample_A", 25, True, 3.14]

print(f"Patient IDs: {patient_ids}")
print(f"Gene Expression Values: {gene_expression_values}")
print(f"Mixed Data: {mixed_data}")
```

**Indexing and Slicing:**

*   **Indexing:** Access individual elements using their position (index). Python uses zero-based indexing, meaning the first element is at index 0.
*   **Slicing:** Access a portion (sub-sequence) of a list using a range of indices.

**Example:**

```python
# List of patient samples, representing different conditions
sample_names = ["Control_1", "Treated_1", "Control_2", "Treated_2", "Control_3"]

# Accessing elements by index
print(f"First sample: {sample_names[0]}") # Control_1
print(f"Third sample: {sample_names[2]}") # Control_2
print(f"Last sample: {sample_names[-1]}") # Control_3 (negative indexing counts from the end)

# Slicing a list
# Get the first two samples
first_two_samples = sample_names[0:2] # or sample_names[:2]
print(f"First two samples: {first_two_samples}") # ['Control_1', 'Treated_1']

# Get samples from the second element to the end
from_second_onwards = sample_names[1:]
print(f"Samples from second onwards: {from_second_onwards}") # ['Treated_1', 'Control_2', 'Treated_2', 'Control_3']

# Get the treated samples (every other, starting from index 1)
treated_samples = sample_names[1::2]
print(f"Treated samples: {treated_samples}") # ['Treated_1', 'Treated_2']

# Reverse a list using slicing
reversed_samples = sample_names[::-1]
print(f"Reversed samples: {reversed_samples}")
```

**Common List Methods:**

Lists come with several built-in methods to modify and query them.

*   `.append(item)`: Adds an item to the end of the list.
*   `.insert(index, item)`: Inserts an item at a specified index.
*   `.remove(item)`: Removes the first occurrence of a specified item.
*   `.pop(index)`: Removes and returns the item at a given index (if no index, removes and returns the last item).
*   `len(list)`: A built-in function (not a method) that returns the number of items in the list.

**Example:**

```python
mutation_list = ["BRAF V600E", "TP53 R175H"]
print(f"Initial mutation list: {mutation_list}")

# Add a new mutation using append()
mutation_list.append("EGFR L858R")
print(f"After append: {mutation_list}")

# Insert a mutation at a specific position (e.g., at the beginning)
mutation_list.insert(0, "KRAS G12D")
print(f"After insert: {mutation_list}")

# Get the length of the list
num_mutations = len(mutation_list)
print(f"Number of mutations: {num_mutations}")

# Remove a specific mutation using remove()
mutation_list.remove("TP53 R175H")
print(f"After remove: {mutation_list}")

# Remove the last mutation using pop()
last_mutation = mutation_list.pop()
print(f"Removed last mutation: {last_mutation}")
print(f"List after pop: {mutation_list}")

# Remove an element at a specific index using pop()
first_mutation = mutation_list.pop(0)
print(f"Removed first mutation: {first_mutation}")
print(f"List after pop(0): {mutation_list}")
```

#### Tuples: Ordered, Immutable Sequences

A **tuple** is similar to a list in that it is an ordered collection of items. However, the key difference is that tuples are *immutable*, meaning their elements cannot be changed, added, or removed after the tuple is created. This immutability makes tuples suitable for data that should not change, such as fixed records or constant parameters.

**Creation:**
Tuples are created by placing all the items inside parentheses `()`, separated by commas. A single item tuple requires a trailing comma.

**Example:**

```python
# A tuple representing a patient's fixed demographic data (ID, DOB, Gender)
patient_record = ("P001", "1985-07-23", "Female")

# A tuple representing a genomic coordinate (chromosome, start, end)
gene_location = ("chr1", 100000, 150000)

# A tuple of enzyme kinetics parameters (Km, Vmax)
enzyme_params = (0.5, 100.2)

# Single-item tuple (note the comma)
single_marker = ("BRCA1",)

print(f"Patient Record: {patient_record}, Type: {type(patient_record)}")
print(f"Gene Location: {gene_location}, Type: {type(gene_location)}")
print(f"Enzyme Parameters: {enzyme_params}, Type: {type(enzyme_params)}")
print(f"Single Marker: {single_marker}, Type: {type(single_marker)}")
```

**Indexing and Slicing:**
Like lists, tuples support indexing and slicing to access elements or sub-sequences.

**Example:**

```python
clinical_trial_info = ("Trial_XYZ", 200, "Phase III", True, "Drug A") # (Trial ID, Participants, Phase, Active, Drug)

# Accessing elements by index
print(f"Trial ID: {clinical_trial_info[0]}")
print(f"Trial Phase: {clinical_trial_info[2]}")

# Slicing a tuple
participants_and_phase = clinical_trial_info[1:3]
print(f"Participants and Phase: {participants_and_phase}")

# Get all info except the first element
remaining_info = clinical_trial_info[1:]
print(f"Remaining Info: {remaining_info}")
```

**When to use Tuples (vs. Lists):**

*   **Immutability:** Use tuples when you want to ensure that the data collection remains constant throughout the program's execution. This can prevent accidental modification of critical data.
*   **Performance:** Tuples are generally slightly faster than lists, especially for fixed-size data. They also consume less memory.
*   **Data Integrity:** Ideal for representing fixed records (e.g., a patient's date of birth and ID, which shouldn't change) or multi-dimensional coordinates.
*   **Dictionary Keys:** Tuples can be used as keys in dictionaries (unlike lists) because they are immutable.
*   **Function Return Values:** Functions often return multiple values as a tuple.

**Example Use Cases in Biomedical Informatics:**

*   Storing constant reference values, like molecular weights or genetic codes.
*   Representing immutable biological entities, such as a specific SNP (Single Nucleotide Polymorphism) defined by (chromosome, position, reference_allele, alternative_allele).
*   Returning multiple metrics from a function (e.g., `(mean_expression, std_dev_expression)`).
*   Database records where certain fields should not be altered.

#### Dictionaries: Unordered Collections of Key-Value Pairs

A **dictionary** is an unordered collection of data values, used to store data in key:value pairs. It is a fundamental data structure for representing structured data, metadata, or relationships between entities. Dictionaries are *mutable*, meaning their contents can be changed after creation.

**Creation:**
Dictionaries are created by placing a comma-separated list of `key: value` pairs inside curly braces `{}`.

**Example:**

```python
# A dictionary storing patient metadata
patient_metadata = {
    "patient_id": "P001",
    "age": 45,
    "diagnosis": "Type 2 Diabetes",
    "medication": ["Metformin", "Insulin"],
    "is_active_trial": True
}

# A dictionary mapping gene symbols to their full names
gene_names = {
    "BRCA1": "Breast Cancer Gene 1",
    "TP53": "Tumor Protein P53",
    "EGFR": "Epidermal Growth Factor Receptor"
}

print(f"Patient Metadata: {patient_metadata}")
print(f"Gene Names: {gene_names}")
```

**Accessing Values:**
You can access values in a dictionary using their corresponding keys. If a key does not exist, it will raise a `KeyError`.

**Example:**

```python
# Accessing patient information
print(f"Patient Diagnosis: {patient_metadata["diagnosis"]}")
print(f"Patient Age: {patient_metadata["age"]}")

# Accessing gene full name
print(f"Full name of BRCA1: {gene_names["BRCA1"]}")

# Using .get() method to access values safely (returns None or a default value if key not found)
print(f"Patient Blood Type (using .get()): {patient_metadata.get("blood_type", "Not Available")}")
print(f"Full name of KRAS (using .get()): {gene_names.get("KRAS", "Gene not found")}")
```

**Adding and Modifying Entries:**

Dictionaries are mutable, allowing you to add new key-value pairs or change the values associated with existing keys.

**Example:**

```python
patient_record = {"id": "P002", "gender": "Male", "age": 60}
print(f"Initial patient record: {patient_record}")

# Add a new entry
patient_record["ethnicity"] = "Asian"
print(f"After adding ethnicity: {patient_record}")

# Modify an existing entry
patient_record["age"] = 61
print(f"After modifying age: {patient_record}")

# Add a new key and assign a list of values (e.g., test results)
patient_record["lab_results"] = {"glucose": 120, "cholesterol": 200}
print(f"After adding lab results: {patient_record}")

# Updating multiple entries or adding new ones using .update()
patient_record.update({"gender": "Prefer not to say", "height_cm": 175})
print(f"After update: {patient_record}")
```

**Common Dictionary Methods:**

*   `.keys()`: Returns a view object that displays a list of all the keys in the dictionary.
*   `.values()`: Returns a view object that displays a list of all the values in the dictionary.
*   `.items()`: Returns a view object that displays a list of a dictionary's key-value tuple pairs.
*   `len(dictionary)`: A built-in function that returns the number of key-value pairs.

**Example:**

```python
patient_data = {
    "name": "Jane Doe",
    "age": 30,
    "disease": "Influenza",
    "treatment_start_date": "2023-01-15"
}

print(f"Keys: {patient_data.keys()}")
print(f"Values: {patient_data.values()}")
print(f"Items: {patient_data.items()}")
print(f"Number of entries: {len(patient_data)}")

# Iterating through keys, values, or items
print("\nIterating through keys:")
for key in patient_data.keys():
    print(key)

print("\nIterating through values:")
for value in patient_data.values():
    print(value)

print("\nIterating through items:")
for key, value in patient_data.items():
    print(f"{key}: {value}")
```

#### Sets: Unordered Collections of Unique Elements

A **set** is an unordered collection of unique, immutable elements. This means that a set cannot have duplicate entries, and the items within a set must be of immutable types (like numbers, strings, or tuples). Sets are particularly useful for operations involving uniqueness, membership testing, and mathematical set operations like union, intersection, and difference.

**Creation:**
Sets are created by placing elements inside curly braces `{}` or by using the `set()` constructor. An empty set must be created using `set()` because `{}` creates an empty dictionary.

**Example:**

```python
# Creating a set of unique patient IDs from a list with duplicates
all_patient_ids = ["P001", "P002", "P001", "P003", "P004", "P002"]
unique_patient_ids = set(all_patient_ids)
print(f"Unique Patient IDs: {unique_patient_ids}") # Order is not guaranteed

# Creating a set of gene symbols related to a disease
disease_genes = {"BRCA1", "TP53", "ATM", "CHEK2"}
print(f"Disease Genes: {disease_genes}")

# Creating an empty set
empty_set = set()
print(f"Empty Set: {empty_set}, Type: {type(empty_set)}")
```

**Adding and Removing Elements:**

*   `.add(item)`: Adds a single element to the set.
*   `.remove(item)`: Removes a specified element. Raises a `KeyError` if the element is not found.
*   `.discard(item)`: Removes a specified element if it is present. Does not raise an error if the element is not found.
*   `.pop()`: Removes and returns an arbitrary element from the set (since sets are unordered, there's no "last" or "first" element).

**Example:**

```python
known_allergens = {"Penicillin", "Latex"}
print(f"Initial allergens: {known_allergens}")

# Add a new allergen
known_allergens.add("Sulfite")
print(f"After adding Sulfite: {known_allergens}")

# Try to add an existing allergen (no change, sets only store unique elements)
known_allergens.add("Penicillin")
print(f"After adding existing Penicillin: {known_allergens}")

# Remove an allergen using remove()
known_allergens.remove("Latex")
print(f"After removing Latex: {known_allergens}")

# Use discard() to remove an element that might or might not be present
known_allergens.discard("Aspirin") # Aspirin is not in the set, no error
print(f"After discarding Aspirin: {known_allergens}")

# Removing an arbitrary element using pop()
popped_allergen = known_allergens.pop()
print(f"Popped allergen: {popped_allergen}")
print(f"Set after pop: {known_allergens}")
```

**Set Operations:**
Sets support mathematical set operations which are highly useful in data analysis, especially for comparing collections of items.

*   **Union (`|` or `.union()`):** Returns a new set containing all unique elements from both sets.
*   **Intersection (`&` or `.intersection()`):** Returns a new set containing only the elements common to both sets.
*   **Difference (`-` or `.difference()`):** Returns a new set containing elements in the first set but not in the second.
*   **Symmetric Difference (`^` or `.symmetric_difference()`):** Returns a new set containing elements that are in either set, but not in both.
*   **Subset (`<=` or `.issubset()`):** Checks if one set is a subset of another.
*   **Superset (`>=` or `.issuperset()`):** Checks if one set is a superset of another.

**Example Use Cases in Biomedical Informatics:**

```python
# Genes associated with Disease A and Disease B
disease_a_genes = {"BRCA1", "TP53", "CDK4", "ATM"}
disease_b_genes = {"TP53", "EGFR", "KRAS", "ATM"}

print(f"Genes for Disease A: {disease_a_genes}")
print(f"Genes for Disease B: {disease_b_genes}")

# 1. Genes associated with EITHER Disease A OR Disease B (Union)
all_related_genes = disease_a_genes.union(disease_b_genes)
print(f"\nGenes related to Disease A or B: {all_related_genes}")
# Alternatively: all_related_genes = disease_a_genes | disease_b_genes

# 2. Genes associated with BOTH Disease A AND Disease B (Intersection)
common_genes = disease_a_genes.intersection(disease_b_genes)
print(f"Genes common to both diseases: {common_genes}")
# Alternatively: common_genes = disease_a_genes & disease_b_genes

# 3. Genes in Disease A but NOT in Disease B (Difference)
disease_a_specific_genes = disease_a_genes.difference(disease_b_genes)
print(f"Genes specific to Disease A: {disease_a_specific_genes}")
# Alternatively: disease_a_specific_genes = disease_a_genes - disease_b_genes

# 4. Genes that are in EITHER set, but NOT in their intersection (Symmetric Difference)
unique_to_one_disease = disease_a_genes.symmetric_difference(disease_b_genes)
print(f"Genes unique to either Disease A or B (not common): {unique_to_one_disease}")
# Alternatively: unique_to_one_disease = disease_a_genes ^ disease_b_genes

# 5. Checking for subsets and supersets
known_oncogenes = {"BRCA1", "TP53"}
print(f"\nAre known oncogenes a subset of Disease A genes? {known_oncogenes.issubset(disease_a_genes)}") # True
print(f"Is Disease A genes a superset of known oncogenes? {disease_a_genes.issuperset(known_oncogenes)}") # True

# Scenario: Comparing two patient cohorts based on their genetic markers
cohort1_markers = {"SNP_A", "SNP_B", "SNP_C"}
cohort2_markers = {"SNP_B", "SNP_D", "SNP_E"}

print(f"\nMarkers in Cohort 1: {cohort1_markers}")
print(f"Markers in Cohort 2: {cohort2_markers}")

# Find markers present in both cohorts
shared_markers = cohort1_markers.intersection(cohort2_markers)
print(f"Shared markers between cohorts: {shared_markers}")

# Find markers unique to Cohort 1
unique_to_cohort1 = cohort1_markers.difference(cohort2_markers)
print(f"Markers unique to Cohort 1: {unique_to_cohort1}")
```
